```python
import matplotlib.pyplot as plt #import modul matplotlib.pyplot sebagai plt
%matplotlib inline

from skimage import data #import data dari skimage
from skimage.io import imread #import imread dari skimage.io
from skimage.color import rgb2gray #import rgb2gray dari skimage.color
import numpy as np #import modul numpy sebagai np
from skimage import io, color #import io color dari skimage
```

```python
citra1 = imread(fname="D:\Semester 6\PCD\cod.jpg") #membaca gambar yang ada di directory tersebut menggunakan fungsi imread dan disimpan pada variabel citra1
citra2 = imread(fname="D:\Semester 6\PCD\lastofus.jpeg") #membaca gambar yang ada di directory tersebut menggunakan fungsi imread dan disimpan pada variabel citra2
grayscale_image = color.rgb2gray(citra1) #mengubah gambar citra1 ke citra grayscale
grayscale_image2 = color.rgb2gray(citra2) #mengubah gambar citra2 ke citra grayscale


print('Shape citra 1 : ', grayscale_image.shape) #mencetak ukuran gambar citra1
print('Shape citra 2 : ', grayscale_image2.shape) #mencetak ukuran gambar citra2
 
fig, axes = plt.subplots(1, 2, figsize=(10, 10)) #membuat objek gambar dan subplot ukuran 1x2 dengan size total 10x10
ax = axes.ravel() #mengakses setiap subplot secara terpisah

ax[0].imshow(grayscale_image, cmap = 'gray') #menampilkan gambar yang ada pada variabel citra1 dengan citra warna gray
ax[0].set_title("Citra 1") #diberi judul Citra 1
ax[1].imshow(grayscale_image2, cmap = 'gray') #menampilkan gambar yang ada pada variabel citra2 dengan citra warna gray
ax[1].set_title("Citra 2") #diberi judul Citra 2
```

```python
copyCitra1 = grayscale_image.copy().astype(float) #mengcopy gambar pada variabel citra1 dengan type jadi float
copyCitra2 = grayscale_image2.copy().astype(float) #mengcopy gambar pada variabel cotra2 dengan type jadi float

m1,n1 = copyCitra1.shape #mengambil nilai baris dan kolom pada hasil copy gambar citra1
output1 = np.empty([m1, n1]) #membuat array kosong 

m2,n2 = copyCitra2.shape #mengambil nilai baris dan kolom pada hasil copy gambar citra2
output2 = np.empty([m2, n2]) #membuat array kosong
print('Shape copy citra 1 : ', copyCitra1.shape) #mencetak Shape copy citra 1 dan menampilkan shape copyCitra1
print('Shape output citra 1 : ', output1.shape) #memcetak Shape output citra 1 dan menampilkan shape output1

print('m1 : ',m1) #mencetak jumlah baris 
print('n1 : ',n1) #mencetak jumlah kolom
print() #mencetak

print('Shape copy citra 2 : ', copyCitra2.shape) #mencetak Shape copy citra 2 dan menampilkan shape copyCitra2
print('Shape output citra 3 : ', output2.shape) #mencetak Shape ouput citra 3 dan menampilkan shape output2
print('m2 : ',m2) #mencetak jumlah baris
print('n2 : ',n2) #mencetak jumlah kolom
print() #mencetak
```

```python
for baris in range(0, m1-1): #melakukan loop pada setiap baris mulai dari 0 sampai m1-1
    for kolom in range(0, n1-1): #melakukan loop pada setiap baris mulai dari 0 sampai n1-1
        a1 = baris #inisialisasi nilai baris pada variabel a1
        b1 = kolom #inisialisasi nilai kolom pada variabel b1
        jumlah = copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1-1] + \
                 copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] + \
                 copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1];  #menghitung jumlah dari pixel tetangga di sekitar
        output1[a1, b1] = (1/9 * jumlah) #menghitung nilai rata-rata
```

```python
for baris1 in range(0, m2-1): #melakukan loop pada setiap baris dari 0 sampai m2-1
    for kolom1 in range(0, n2-1): #melakukan loop pada setiap kolom dari 0 sampai n2-1
        a1 = baris1 #inisialisasi variabel a1 menjadi nilai baris1
        b1 = kolom1 #inisialisasi variabel b1 menjadi nilai kolom1
        jumlah = copyCitra2[a1-1, b1-1] + copyCitra2[a1-1, b1] + copyCitra2[a1-1, b1-1] + \
                 copyCitra2[a1, b1-1] + copyCitra2[a1, b1] + copyCitra2[a1, b1+1] + \
                 copyCitra2[a1+1, b1-1] + copyCitra2[a1+1, b1] + copyCitra2[a1+1, b1+1];  #menghitung jumlah dari pixel tetangga di sekitar
        output2[a1, b1] = (1/9 * jumlah) #menghitung nilai rata-rata
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10)) #membuat objek gambar dan subplot ukuran 2x2 dengan size total 10x10
ax = axes.ravel() #mengakses setiap subplot secara terpisah

ax[0].imshow(grayscale_image, cmap = 'gray') #menampilkan gambar pada subplot pertama  
ax[0].set_title("Input Citra 1") #diberi judul Input Citra 1
 
ax[1].imshow(grayscale_image2, cmap = 'gray') #menampilkan gambar pada subplot kedua
ax[1].set_title("Input Citra 2") #diberi judul Input Citra 2

ax[2].imshow(output1, cmap = 'gray') #menampilkan gambar variabel output1 pada subplot ketiga
ax[2].set_title("Output Citra 1") #diberi judul Output Citra 1

ax[3].imshow(output2, cmap = 'gray') #menampilkan gambar variabel output2 pada subplot keempat 
ax[3].set_title("Output Citra 2") #diberi judul Output Citra 2
```
