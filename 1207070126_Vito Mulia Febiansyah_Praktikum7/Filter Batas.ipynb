```python
import matplotlib.pyplot as plt #import modul matplotlib.pyplot sebagai plt
%matplotlib inline

from skimage import data #import data dari skimage
from skimage.io import imread #import imread dari skimage.io
from skimage.color import rgb2gray  #import rgb2gray dari skimage.color
import numpy as np #import numpy sebgai np
from skimage import io, color #import io color dari skimage
```

```python
citra1 = imread(fname="D:\Semester 6\PCD\cod.jpg") #membaca gambar menggunakan imread dan disimpan pada variabel citra1
citra2 = imread(fname="D:\Semester 6\PCD\lastofus.jpeg") #membaca gambar menggunakan imread dan disimpan pada variabel citra2
grayscale_image = color.rgb2gray(citra1) #mengubah gambar citra1 ke citra grayscale
grayscale_image2 = color.rgb2gray(citra2) #mengubah gambar citra2 ke citra grayscale

print('Shape citra 1 : ', citra1.shape) #mencetak shape citra1
print('Shape citra 2 : ', citra2.shape) #mencetak shape citra2

fig, axes = plt.subplots(1, 2, figsize=(10, 10)) #membuat objek gambar dan subplot ukuran 1x2 dengan size total 10x10
ax = axes.ravel() #mengakses setiap subplot secara terpisah

ax[0].imshow(grayscale_image, cmap = 'gray') #menampilkan gambar citra1 dengan warna citra gray
ax[0].set_title("Citra 1") #diberi judul Citra 1
ax[1].imshow(grayscale_image2, cmap = 'gray') #menampilkan gambar citra2 dengan warna citra gray
ax[1].set_title("Citra 2") #diberi judul Citra 2
```

```python
copyCitra1 = grayscale_image.copy() #mengcopy gambar yang ada pada variabel grayscale_image dan disimpan pada copyCitra1
copyCitra2 = grayscale_image2.copy() #mengcopy gambar yang ada pada variabel grayscale_image2 dan disimpan pada copyCitra2

m1,n1 = copyCitra1.shape #mengambil nilai baris dan kolom copyCitra1
output1 = np.empty([m1, n1]) #membuat array kosong

m2,n2 = copyCitra2.shape #mengambil nilai baris dan kolom copyCitra2
output2 = np.empty([m2, n2]) #membuat array kosong
print('Shape copy citra 1 : ', copyCitra1.shape) #mencetak shape copyCitra1
print('Shape output citra 1 : ', output1.shape) #mencetak shape output copyCitra1

print('m1 : ',m1) #mencetak baris copyCitra1
print('n1 : ',n1) #mencetak kolom copyCitra1
print() #mencetak

print('Shape copy citra 2 : ', copyCitra2.shape) #mencetak shape copyCitra2
print('Shape output citra 2 : ', output2.shape) #mencetak shape output copyCitra2
print('m2 : ',m2) #mencetak baris copyCitra2
print('n2 : ',n2) #mencetak kolom copyCitra2
print() #mencetak
```

```python
for baris in range(0, m1-1): #melakukan loop pada setiap baris dari 0 sampai m1-1
    for kolom in range(0, n1-1): #melakukan loop pada setiap kolom dari 0 sampai n1-1
        
        a1 = baris #inisialisasi nilai baris pada variabel a1
        b1 = kolom #inisialisasi nilai kolom pada variabel b1
        
        arr = np.array([copyCitra1[a1-1, b1-1], copyCitra1[a1-1, b1], copyCitra1[a1, b1+1], \
            copyCitra1[a1, b1-1], copyCitra1[a1, b1+1], copyCitra1[a1+1, b1-1],  \
            copyCitra1[a1+1, b1], copyCitra1[a1+1, b1+1]]) #menghitung jumlah dari pixel tetangga di sekitar
        
        minPiksel = np.amin(arr); #menghitung nilai pixel terendah         
        maksPiksel = np.amax(arr); #menghitung nilai pixel tertinggi
            
        if copyCitra1[baris, kolom] < minPiksel : #operasi looping dengan baris dan kolom
            output1[baris, kolom] = minPiksel #jika nilai pixel kurang dari minPiksel, maka nilainya akan minPiksel
        else : 
            if copyCitra1[baris, kolom] > maksPiksel : #jika lebih dari maksPiksel
                output1[baris, kolom] = maksPiksel #maka nilainya akan maksPiksel
            else : 
                output1[baris, kolom] = copyCitra1[baris, kolom] #nilai pixel output1 sama dengan nilai copyCitra1
```

```python
for baris1 in range(0, m2-1): #melakukan loop pada setiap baris1 dari 0 sampai m2-1
    for kolom1 in range(0, n2-1): #melakukan loop pada setiap kolom1 dari 0 sampai n2-1
        
        a1 = baris1 #inisialisasi nilai baris1 pada variabel a1
        b1 = kolom1 #inisialisasi nilai kolom1 pada variabel b1
        
        arr = np.array([copyCitra2[a1-1, b1-1], copyCitra2[a1-1, b1], copyCitra2[a1, b1+1], \
            copyCitra2[a1, b1-1], copyCitra2[a1, b1+1], copyCitra2[a1+1, b1-1],  \
            copyCitra2[a1+1, b1], copyCitra2[a1+1, b1+1]]) #menghitung jumlah dari pixel tetangga di sekitar
        
        minPiksel = np.amin(arr); #menghitung nilai pixel terendah        
        maksPiksel = np.amax(arr); #menghitung nilai pixel tertinggi
            
        if copyCitra2[baris1, kolom1] < minPiksel : #operasi loop, jika kurang dari minPiksel
            output2[baris1, kolom1] = minPiksel #maka nilainya akan jadi minPiksel
        else :
            if copyCitra2[baris1, kolom1] > maksPiksel : #jika lebih dari maksPiksel
                output2[baris1, kolom1] = maksPiksel #maka nilainya akan jadi maksPiksel
            else :
                output2[baris1, kolom1] = copyCitra2[baris1, kolom1] #nilai pixel output2 sama dengan nilai copyCitra2
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10)) #membuat objek gambar dan subplot ukuran 2x2 dengan size total 10x10
ax = axes.ravel() #mengakses setiap subplot secara terpisah

ax[0].imshow(grayscale_image, cmap = 'gray') #menampilkan gambar pada subplot pertama dengan citra gray
ax[0].set_title("Input Citra 1") #diberi judul Input Citra 1

ax[1].imshow(grayscale_image2, cmap = 'gray') #menampilkan gambar pada subplot kedua dengan citra gray
ax[1].set_title("Input Citra 2") #diberi judul Input Citra 2

ax[2].imshow(output1, cmap = 'gray') #menampilkan gambar pada subplot ketiga dengan citra gray
ax[2].set_title("Output Citra 1") #diberi judul Output Citra 1

ax[3].imshow(output2, cmap = 'gray') #menampilkan gambar pada subplot keempat dengan citra gray
ax[3].set_title("Output Citra 2") #diberi judul Output Citra 2
```
