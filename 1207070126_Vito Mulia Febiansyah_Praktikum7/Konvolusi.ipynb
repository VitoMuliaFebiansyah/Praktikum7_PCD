```python
import matplotlib.pyplot as plt #import modul matplotlib.pyplot sebagai plt
%matplotlib inline

from skimage import data #import data dari skimage
from skimage.io import imread #import imread dari skimage.io
from skimage.color import rgb2gray #import rgb2gray dari skimage.color
import numpy as np #import modul numpy sebagai np

import cv2 #import modul cv2
```

```python
citra1 = imread(fname="D:\Semester 6\PCD\cod.jpg") #membaca gambar mobil menggunakan imread dan disimpan pada variabel citra1
print(citra1.shape) #mencetak nilai shape gambar yang ada pada citra1

plt.imshow(citra1, cmap='gray') #menampilkan gambar citra1 dengan citra gray
```

```python
kernel = np.array([[-1, 0, -1], 
                   [0, 4, 0], 
                   [-1, 0, -1]]) #matriks 3x3 yang disimpan pada kernel untuk digunakan proses filtering

citraOutput = cv2.filter2D(citra1, -1, kernel) #konvolusi antara citra1 dan kernel

fig, axes = plt.subplots(1, 2, figsize=(12, 12)) #membuat objek gambar dan subplot ukuran 1x2 dengan size total 12x12
ax = axes.ravel() #mengakses setiap subplot secara terpisah

ax[0].imshow(citra1, cmap = 'gray') #menampilkan gambar citra 1 dengan citra gray
ax[0].set_title("Citra Input") #diberi nama Citra Input
ax[1].imshow(citraOutput, cmap = 'gray') #menampilkan gambar citra output dengan citra gray
ax[1].set_title("Citra Output") #diberi nama Citra Output
```
