```python
import cv2 #import modul cv2
import numpy as np #import modul numpy sebagai np
from skimage import data #import data dari skimage
import matplotlib.pyplot as plt #import modul matplotlib.pyplot sebagai plt
%matplotlib inline
```

```python
img = data.moon() #mengambil data gambar moon dan disimpan pada variabel img

row, column = img.shape #mengambil nilai baris dan kolom pada gambar

img1 = np.zeros((row,column),dtype = 'uint8') #membuat array kosong dengan type uint8 dan disimpan pada variabel img1
 

min_range = 10 #inisialisasi nilai 10 untuk variabel min_range
max_range = 60 #inisialisasi nilai 60 untuk variabel max_range
 

for i in range(row): #melakukan loop pada setiap baris
    for j in range(column): #melakukan loop pada setiap kolom
        if img[i,j]>min_range and img[i,j]<max_range: #jika baris kolom gambar lebih dari min_range dan baris kolom gambar kurang dari max_range
            img1[i,j] = 255 #jika lebih maka nilainya 255
        else:
            img1[i,j] = 0 #jika kurang maka nilainya 0
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12)) #membuat objek gambar dan subplot ukuran 2x2 dengan size total 12x12
ax = axes.ravel() #mengakses setiap subplot secara terpisah

ax[0].imshow(img, cmap=plt.cm.gray) #menampilkan gambar pada subplot pertama dengan citra keabuan
ax[0].set_title("Citra Input") #diberi judul Citra Input
ax[1].hist(img.ravel(), bins=256) #membuat histogram gambar pada subplot kedua dengan bin 256
ax[1].set_title('Histogram Input') #diberi judul Histogram Input

ax[2].imshow(img1, cmap=plt.cm.gray) #menampilkan gambar pada subplot ketiga dengan citra keabuan
ax[2].set_title("Citra Output") #diberi judul Citra Output 
ax[3].hist(img1.ravel(), bins=256) #membuat histogram pada subplot keempat dengan bin 256
ax[3].set_title('Histogram Output') #diberi judul Histogram Output
```
